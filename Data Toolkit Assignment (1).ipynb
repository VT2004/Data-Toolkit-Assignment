Q3.Explain the following terms : The difference in np.array, np.asarray and np.asanyarray


 The difference between Deep copy and shallow copy

np.array():
 Creates a new NumPy array. If the input is already an array, it creates a copy.


```python
#This will always create a new array, even if the input is already an array.
arr = np.array([1, 2, 3])
arr
```

np.asarray():
 Converts input to a NumPy array. If the input is already a NumPy array, it returns the same array without copying.

```python
#This is efficient when you want to ensure the output is a NumPy array but avoid unnecessary copying.
arr1 = np.asarray([1, 2, 3])
arr1
```

np.asanyarray():
Similar to np.asarray(), but it preserves array subclasses (e.g., np.matrix).

```python
#This is useful when working with special types of arrays and you want to retain their type.
matrix = np.matrix([[1, 2], [3, 4]])
arr2 = np.asanyarray(matrix)
arr2
```

Shallow Copy:
Creates a new object, but nested objects (e.g., lists within lists) are not copied; they are referenced. Changes to nested objects affect both the original and the copied object.


Deep Copy:
Creates a new object and recursively copies all nested objects, making them independent. Changes to nested objects in the copy do not affect the original.

Q13. Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:

 a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.

b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.

c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.

 d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.

 e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.

 f) Display a legend for the scatter plot, the horizontal line, and the vertical line.

```python
# Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1
import matplotlib.pyplot as plt
x = np.random.rand(100)
y = np.random.rand(100)

#  Create a scatter plot
plt.scatter(x, y, color='red', marker='o')

# Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.
plt.axhline(y=0.5, color='blue', linestyle='--', label='y = 0.5')

# Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.
plt.axvline(x=0.5, color='green', linestyle=':', label='x = 0')

# Label the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Advanced Scatter Plot with Lines')

# Display a legend for the scatter plot, horizontal line, and vertical line
plt.legend()
plt.show()
```

Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

 b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
np.random.seed(0)
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
temperature = np.random.uniform(20, 30, 100)
humidity = np.random.uniform(40, 60, 100)

df = df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

fig = plt.figure(figsize=(12, 6))

ax1 = plt.subplot(111)
ax1.plot(df['Date'], df['Temperature'], 'b-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'r-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='r')
ax2.tick_params('y', colors='r')

plt.title('Temperature and Humidity Over Time')

plt.show()
```

Q15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

a) Plot a histogram of the data with 30 bins.

 b) Overlay a line plot representing the normal distribution's pprobability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.


```python
import scipy.stats as stats

data = np.random.normal(0, 1, 1000)

plt.figure()

plt.hist(data, bins=30, color='blue', alpha=0.7, label='Histogram')
x = np.linspace(np.mean(data), np.std(data), 100)
pdf = stats.norm.pdf(x, loc=np.mean(data), scale=np.std(data))

plt.plot(x, pdf, 'r-', lw=2, label='Normal Distribution PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

Q17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

df = pd.DataFrame({'x': x, 'y': y})
df
```

```python
import seaborn as sns

def determine_quadrant(row):
    if row['x'] > 0 and row['y'] > 0:
        return 'Quadrant 1'
    elif row['x'] < 0 and row['y'] > 0:
        return 'Quadrant 2'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant 3'
    elif row['x'] > 0 and row['y'] < 0:
        return 'Quadrant 4'
    else:
        return 'Origin'

df['Quadrant'] = df.apply(determine_quadrant, axis=1)

fig = plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')
plt.legend(title='Quadrant', loc='upper left')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.show()
```

Q18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="Sine Wave", line_width=2)
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
show(p)
```

Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart’.

```python
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import HoverTool


categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(10, 100, 10)

df = pd.DataFrame({'Category': categories, 'Value': values})
source = ColumnDataSource(df)

p = figure(x_range=df['Category'], title='Random Categorical Bar Chart', toolbar_location=None, tools="")

mapper = linear_cmap(field_name='Value', low=min(values), palette = 'Viridis256', high=max(values))

p.vbar(x='Category', top='Value', width=0.9, source=source, line_color='white', fill_color=mapper)

hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'

show(p)
```

Q20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot’

```python
import plotly.graph_objects as go
x = np.arange(10)
y = np.random.random(10)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Random Data'))

fig.update_layout(
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    title='Simple Line Plot'
)

fig.show()
```
